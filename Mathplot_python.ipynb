3. Write a Python program to draw a line using given axis values taken from a text file, with
suitable label in the x axis, y axis and a title.
Test Data:
test.txt
1 2
2 4
3 1

```python
import matplotlib.pyplot as plt

x_values = []
y_values = []

with open("C:\\Users\\Vamsi\\OneDrive\\Desktop\\test.txt",'r') as file:
    for line in file:
        x, y = map(int, line.split())
        x_values.append(x)
        y_values.append(y)

plt.plot(x_values, y_values, marker='o', linestyle='-.', color='b', label="Line Graph")

plt.xlabel("X-Axis Label")
plt.ylabel("Y-Axis Label")
plt.title("Line Graph from test.txt")
plt.legend()

plt.show()
```

5. Write a Python program to plot two or more lines on same plot with suitable legends of each
line.

```python
import matplotlib.pyplot as plt
import numpy as np
x=np.array([1,4,8,10])
y1=np.array([1,5,6,8])
y2=np.array([2,4,6,8])
plt.plot(x,y1,"*",linestyle="-",color="g",label="graph1")
plt.plot(x,y2,"*",linestyle="-",color="g",label="graph2")
plt.legend()
plt.show()
```

6. Write a Python program to plot two or more lines with legends, different widths and colors

9. Write a Python program to display the current axis limits values and set new axis values

```python
x = [1, 2, 3, 4, 5]
y = [10, 20, 25, 30, 40]

plt.plot(x, y)

current_x_limits = plt.gca().get_xlim()
current_y_limits = plt.gca().get_ylim()
print(f"Current X-axis limits: {current_x_limits}")
print(f"Current Y-axis limits: {current_y_limits}")

plt.xlim(0, 6)
plt.ylim(0, 50)

plt.title("Plot with Updated Axis Limits")
plt.xlabel("X-Axis")
plt.ylabel("Y-Axis")
plt.show()
```

12. Write a Python program to create multiple types of charts

```python
x = [1, 2, 3, 4, 5]
y = [10, 20, 25, 30, 40]
categories = ['A', 'B', 'C', 'D', 'E']
values = [5, 10, 15, 20, 25]


plt.subplot(231)
plt.plot(x, y, label="Line Plot", color='blue', marker='o')
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("Line Chart")
plt.legend()

plt.subplot(232)
plt.bar(categories, values, color='green')
plt.xlabel("Categories")
plt.ylabel("Values")
plt.title("Bar Chart")

plt.subplot(233)
plt.scatter(x, y, color='red', label='Scatter Plot')
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("Scatter Plot")
plt.legend()


plt.subplot(234)
sizes = [15, 30, 45, 10]
labels = ['A', 'B', 'C', 'D']
plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=90)
plt.title("Pie Chart")

plt.subplot(235)
data = [5, 10, 15, 10, 5, 15, 20, 20, 25, 25, 30, 30]
plt.hist(data, bins=5, color='purple', edgecolor='black')
plt.title("Histogram")
plt.xlabel("Bins")
plt.ylabel("Frequency")

plt.tight_layout()
plt.show()
```

14. Write a Python program to display the grid and draw line charts of the closing value of
Alphabet Inc. between October 3, 2016 to October 7, 2016. Customized the grid lines with
rendering with a larger grid (major grid) and a smaller grid (minor grid).Turn on the grid but turn
off ticks

```python
import matplotlib.pyplot as plt

dates = ['03-10-16', '04-10-16', '05-10-16', '06-10-16', '07-10-16']
closing_values = [772.559998, 776.429993, 776.469971, 776.859985, 775.080017]

plt.plot(dates, closing_values, label="Closing Value", color='green', marker='o', linestyle='-', linewidth=2)

plt.xlabel("Date")
plt.ylabel("Closing Value ($)")
plt.title("Alphabet Inc. Closing Values (Oct 3-7, 2016)")

plt.grid(True, which='both', axis='both', linestyle='-', linewidth=0.5, color='blue')
plt.minorticks_on()
plt.grid(True, which='major', axis='both', linestyle='-', linewidth=1, color='blue')
plt.grid(True, which='minor', axis='both', linestyle=':', linewidth=0.5, color='red')

plt.tick_params(axis='both', which='both', length=0)

plt.legend()
plt.xticks(rotation=45)
plt.tight_layout()

plt.show()
```

15.Write a Python program to create multiple plots

```python
x = [1, 2, 3, 4, 5]
y1 = [10, 20, 25, 30, 40]
y2 = [5, 10, 15, 20, 25]
y3 = [1, 4, 9, 16, 25]

plt.subplot(2, 2, 1)
plt.plot(x, y1, 'r-', label='y1 = 10x')
plt.title("Plot 1: y = 10x")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()

plt.subplot(2, 2, 2)
plt.plot(x, y2, 'g--', label='y2 = 5x')
plt.title("Plot 2: y = 5x")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()

plt.subplot(2, 2, 3)
plt.plot(x, y3, 'b:', label='y3 = x^2')
plt.title("Plot 3: y = x^2")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()

plt.subplot(2, 2, 4)
plt.plot(x, y1, 'm-', label='y1 = 10x')
plt.plot(x, y2, 'y--', label='y2 = 5x')
plt.title("Plot 4: y = 10x and y = 5x")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()

plt.tight_layout()
plt.show()
```
